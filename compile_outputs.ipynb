```python
import pickle
import pandas as pd
import numpy as np
import random
import os

outPath = '/nobackup/jakravit/data/EAP_batch_outputs/'
species = os.listdir(outPath)

data0 = {'Chlorophyceae': {},
         'Cryptophyceae': {},
         'Coscinodiscophyceae': {},
         'Bacillariophyceae': {},
         'Dinophyceae':{},
         'Prasinophyceae':{},
         'Pelagophyceae':{},
         'Haptophyceae':{},
         'Prymnesiophyceae':{},
         'Raphidophyceae':{},
         'Eustigmatophyceae':{},
         'Cyano_blue':{},
         'Cyano_red':{},
         'Rhodophyceae':{},
         'Synechococcus':{},
         'Prochlorococcus':{}
        }


for k in species:
    with open(f'{outPath}{k}', 'rb') as f:
        speciesdata = pickle.load(f)
        sp = k.split('.')[0]
        if sp == 'M_aeruginosa':
            pft = 'Cyano_blue'
            data0[pft][sp] = speciesdata
        else:
            pft = speciesdata[sp][random.choice(list(speciesdata[sp]))]['PFT1']
            print (sp, pft)
            data0[pft][sp] = speciesdata[sp]
```

```python
with open('/nobackup/jakravit/data/EAP_phytoplankton_dataset/PFTs_1nm_v1103.p', 'wb') as f:
    pickle.dump(data0,f)
```
